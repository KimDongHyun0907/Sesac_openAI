```python
folder = r"C:\Users\bluecom014\Desktop\  \ \img"

from glob import glob
from PIL import Image
import numpy as np

fileList = glob(folder+'\\*')
imgList = []

for f in fileList:
    file_img = Image.open(f).resize((100,200))
    imgArr5 = np.array(file_img)
    imgList.append(imgArr5)

imgList = np.array(imgList)
imgList.shape
```

```python
import matplotlib.pyplot as plt

for cnt, i in enumerate(imgList):
    plt.subplot(1, len(imgList), cnt+1)
    plt.imshow(i)
```

```python
col_size = 2
for idx, f in enumerate(imgList):
    plt.subplot(col_size, len(imgList), idx*2+1)
    plt.imshow(f)
```

```python
imgList2 = imgList
imgList2[:,:50,:,:]=0
```

```python
for cnt, i in enumerate(imgList2):
    plt.subplot(1, len(imgList2), cnt+1)
    plt.imshow(i)
```
