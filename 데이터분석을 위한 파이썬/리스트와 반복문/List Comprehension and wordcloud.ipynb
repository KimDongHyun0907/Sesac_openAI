```python
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud

speech = ['''Ladies and gentlemen, honored guests, and fellow citizens,

Today, we gather here to reflect on our shared values and aspirations. In the face of challenges, let us remember the strength that unites us as a community. Together, we can overcome obstacles and build a brighter future for generations to come.

As we navigate the complexities of our time, let justice, compassion, and understanding guide our actions. It is through collaboration and empathy that we can create a world that celebrates diversity and fosters inclusivity.

In the words of a wise person, 'The only limit to our realization of tomorrow will be our doubts of today.' Let us embrace change with optimism and determination, knowing that together, we can shape a world that reflects the best of humanity.

Thank you for your commitment to a better tomorrow. Let us move forward with courage and conviction, united in our pursuit of a more just and harmonious world.

Thank you.''','alksdf fdfksj asdf qeiuro zovxhi', 'vsfpoqi vxcvo qdf']


for speech_text in speech:
    text = re.sub(r'[^\w\s]', '', speech_text)
    # words = [word for word in text.lower().split() if len(word)>=4]
    # words = ' '.join(words)
    # WordCloud 객체 생성
    # wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(words)
    wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(text)

    # 워드 클라우드를 시각화
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  # 축 제거
    plt.show()
```
