```python
def mySum(a):
    sum=0
    for i in a:
        sum+=i
    return sum

a=[1,2,3,4]
print(mySum(a))
```

```python
def mySum(*a):
    sum=0
    for i in a:
        sum+=i
    return sum

a=[1,2,3,4]
print(mySum(1,2,3,4))
```

```python
from PIL import Image
from glob import glob
import matplotlib.pyplot as plt
import numpy as np

fileList = glob("C:/Users/bluecom014/Desktop/데이터분석을 위한 파이썬/파이썬 함수/img/*.jpg")
print(len(fileList), '개 파일')

for cnt, file in enumerate(fileList):
    # print(file)
    # print('-'*50)

    img = Image.open(file).resize((200,300))
    imgArr = np.array(img)
    # print(imgArr)
    plt.imshow(imgArr+100)
    plt.show()
```

```python
# 파일 불러오기 함수
def fileOpen(fileList):
    xData, yData = [], []
    for file in fileList:
        img = Image.open(file).resize((200,300))
        img = np.array(img)
        xData.append(img)
        yData.append('flower')
    return xData, yData

# 이미지 출력함수
def imgPltShow(num=0, **kwargs):
    for xD, yD in zip(kwargs['x'], kwargs['y']):
        plt.imshow(xD+num)
        plt.title(yD)
        plt.show()
```

```python
from glob import glob

fileList = glob("C:/Users/bluecom014/Desktop/데이터분석을 위한 파이썬/파이썬 함수/img/*.jpg")
img, label = fileOpen(fileList)
imgPltShow(-50, x=img, y=label)
```
