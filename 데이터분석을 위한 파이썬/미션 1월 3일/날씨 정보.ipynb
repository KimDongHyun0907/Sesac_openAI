```python
import json

with open(r'C:\Users\bluecom014\Desktop\데이터분석을 위한 파이썬\파이썬 클래스\수정_202301월자료_weatherData1.json', 'r') as f:
    json_data = json.load(f)

print(json.dumps(json_data, ensure_ascii=False))
```

```python
location = json_data['j']
dates = json_data['d']
min_temp = json_data['max']
max_temp = json_data['min']
rainfall = json_data['rain']
```

```python
dates
```

```python
data = [[j, d, m, n, r] for j, d, m, n, r in zip(location, dates, min_temp, max_temp, rainfall)]

print(data)
```

```python
seoul_data, busan_data, jeju_data = [], [], []

for x in data:
    if x[0] == '서울':
        seoul_data.append(x)
    elif x[0] == '부산':
        busan_data.append(x)
    else:
        jeju_data.append(x)
```

서울, 부산, 제주의 강수량 합 구하기

```python
def total_rainfall(location_list, ans):
    for _, _, _, _, r in location_list:
        if r != '':
            ans += float(r)

    return ans

seoul_total_rainfall, busan_total_rainfall, jeju_total_rainfall = 0, 0, 0


seoul_total_rainfall = total_rainfall(seoul_data, seoul_total_rainfall)
busan_total_rainfall = total_rainfall(busan_data, busan_total_rainfall)
jeju_total_rainfall = total_rainfall(jeju_data, jeju_total_rainfall)

print(f'서울의 강수량 합은 {seoul_total_rainfall:.1f}')
print(f'부산의 강수량 합은 {busan_total_rainfall:.1f}')
print(f'제주의 강수량 합은 {jeju_total_rainfall:.1f}')
```

서울, 부산, 제주 강수량 차트 출력

```python
import matplotlib.pyplot as plt

def make_chart(idx, rainfall_list, title, min_val = 0, max_val = 80):
    plt.subplot(1, 3, idx)
    plt.plot(rainfall_list)
    plt.title(title)
    plt.ylim([min_val, max_val])

plt.rc('font', family = 'NanumBarunGothic')

seoul_rainfall = [float(x[4]) if x[4] != '' else 0.0 for x in seoul_data]
busan_rainfall = [float(x[4]) if x[4] != '' else 0.0 for x in busan_data]
jeju_rainfall = [float(x[4]) if x[4] != '' else 0.0 for x in jeju_data]

plt.figure(figsize=(20, 5))

make_chart(1, seoul_rainfall, '서울 강수량', 0, 50)
make_chart(2, busan_rainfall, '부산 강수량', 0, 100)
make_chart(3, jeju_rainfall, '제주 강수량', 0, 50)
```

```python
def make_temp_chart(idx, temp_list_max, temp_max_text, temp_list_min, temp_min_text, chart_title):
    plt.subplot(1, 3, idx)
    plt.plot(temp_list_max, marker='o', label = temp_max_text)
    plt.plot(temp_list_min, marker='o', label = temp_min_text)
    plt.title(chart_title)

seoul_temp_max = [float(x[2]) for x in seoul_data]
seoul_temp_min = [float(x[3]) for x in seoul_data]

busan_temp_max = [float(x[2]) for x in busan_data]
busan_temp_min = [float(x[3]) for x in busan_data]

jeju_temp_max = [float(x[2]) for x in jeju_data]
jeju_temp_min = [float(x[3]) for x in jeju_data]

plt.figure(figsize=(20, 5))

make_temp_chart(1, seoul_temp_max, '서울 최고 기온', seoul_temp_min, '서울 최저 기온', '서울 기온')
make_temp_chart(2, busan_temp_max, '부산 최고 기온', busan_temp_min, '부산 최저 기온', '부산 기온')
make_temp_chart(3, jeju_temp_max, '제주 최고 기온', jeju_temp_min, '제주 최저 기온', '제주 기온')
```
