```python
print('<<1~30 숫자 맞히기 게임>>')

ans = np.random.randint(1, 31)

while True:
    num = int(input('숫자 입력 (종료 0): '))
    if num == 0:
        break
    if num < ans:
        print('더 큰 숫자 입력!')
    elif num > ans:
        print('더 작은 숫자 입력!')
    else:
        print('정답!!')
        break
    
```

```python
salary = [200, 150, 180, 220, 250]
new_salary = np.array(salary, dtype='float64')*1.5
print('인상된 금액: ', new_salary)
avg_salary = new_salary.mean()
print('평균 금액: ', avg_salary)
result = new_salary >= avg_salary
print('평균 이상인 원소: ', new_salary[result])
print('평균이상인 원소의 개수: ', len(new_salary[result]))
```
