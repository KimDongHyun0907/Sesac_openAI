```python
import cv2
import time
import datetime

# 얼굴 인식을 위한 Haarcascades 분류기 경로
face_cascade_path = 'haarcascade_frontalface_default.xml'

# Haarcascades 분류기 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + face_cascade_path)

# 웹캠 열기
cap = cv2.VideoCapture(0)
cnt = 0

while True:
    # 프레임 읽기
    ret, frame = cap.read()

    # 회색조로 변환
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 감지
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

    # 감지된 얼굴에 사각형 그리기
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # 화면에 표시
    cv2.imshow('Face Detection', frame)

    # 얼굴이 감지되면 3초 후에 사진을 촬영
    if len(faces) > 0:
        # time.sleep(3)
        print("얼굴이 감지되었습니다. 3초 동안 카운트다운을 시작합니다.")
        for i in range(3, 0, -1):
            cv2.putText(frame, str(i), (int(frame.shape[1] / 2) - 10, int(frame.shape[0] / 2) + 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 3)
            cv2.imshow('Face Detection', frame)
            cv2.waitKey(1000)  # 1초 대기
            frame = cap.read()[1]  # 다음 프레임 읽기

        print("사진을 촬영합니다.")
        ret, frame = cap.read()
        d = datetime.datetime.now()
        d_list = [d.year, d.month, d.day, d.hour, d.minute, d.second]
        date_name = '_'.join(map(str, d_list))
        cv2.imwrite('picture/captured_face'+str(cnt)+'_'+date_name+'.jpg', frame)
        cnt += 1
        print("사진이 촬영되었습니다.")

    # 종료 키 설정 (ESC 키)
    if cv2.waitKey(1) == 27:
        break

# 리소스 해제
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

def detect_faces():
    # 웹캠을 켭니다.
    cap = cv2.VideoCapture(0)

    # 얼굴 인식을 위해 얼굴 검출기를 로드합니다.
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # 다른 이미지를 불러옵니다.
    overlay_image = cv2.imread('ironman.png')

    while True:
        # 웹캠으로부터 영상을 받아옵니다.
        ret, frame = cap.read()

        if not ret:
            break

        # 영상을 회색으로 변환합니다.
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # 얼굴을 검출합니다.
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # 검출된 얼굴 주위에 사각형을 그리고 이미지를 합성합니다.
        for (x, y, w, h) in faces:
            # 얼굴에 이미지 합성
            resized_overlay = cv2.resize(overlay_image, (w, h))
            
            # 알파 채널이 없을 경우, 알파 채널을 추가합니다.
            if resized_overlay.shape[2] == 3:
                resized_overlay = cv2.cvtColor(resized_overlay, cv2.COLOR_BGR2BGRA)

            frame[y:y+h, x:x+w] = frame[y:y+h, x:x+w] * (1 - resized_overlay[:, :, 3:] / 255.0) + \
                                   resized_overlay[:, :, :3] * (resized_overlay[:, :, 3:] / 255.0)

            # 얼굴 주위에 사각형 그리기
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # 결과 영상을 화면에 출력합니다.
        cv2.imshow('Face Detection', frame)

        # 'q'를 누르면 종료합니다.
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # 웹캠을 끕니다.
    cap.release()
    cv2.destroyAllWindows()

# 웹캠으로부터 얼굴을 인식하고 다른 이미지를 합성합니다.
detect_faces()
```

```python
import cv2
import numpy as np

# 얼굴과 눈을 감지하는 Haar Cascade 분류기를 불러옴
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# 눈에 합성할 별 이미지를 불러옴
star_img = cv2.imread('star.png', -1)

# 웹캠 열기
cap = cv2.VideoCapture(0)

while True:
    # 프레임 읽기
    ret, frame = cap.read()

    # 그레이스케일로 변환
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 검출
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # 얼굴 주위에 사각형 그리기
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # 눈 검출
        roi_gray = gray[y:y + h, x:x + w]
        eyes = eye_cascade.detectMultiScale(roi_gray)

        for (ex, ey, ew, eh) in eyes:
            # 눈 주위에 사각형 그리기
            cv2.rectangle(frame, (x + ex, y + ey), (x + ex + ew, y + ey + eh), (0, 255, 0), 2)

            # 별 이미지 크기 조절
            resized_star = cv2.resize(star_img, (ew, eh))

            # 눈에 별 합성
            for c in range(0, 3):
                if resized_star.shape[2] == 4:  # 별 이미지에 알파 채널이 있는 경우
                    frame[y + ey:y + ey + eh, x + ex:x + ex + ew, c] = frame[y + ey:y + ey + eh, x + ex:x + ex + ew, c] * (
                            1 - resized_star[:, :, 3] / 255.0) + resized_star[:, :, c] * (resized_star[:, :, 3] / 255.0)
                else:  # 별 이미지에 알파 채널이 없는 경우
                    frame[y + ey:y + ey + eh, x + ex:x + ex + ew, c] = resized_star[:, :, c]

    # 화면에 출력
    cv2.imshow('Eye with Star', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 웹캠 해제
cap.release()
cv2.destroyAllWindows()
```
