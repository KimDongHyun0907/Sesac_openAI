```python
import pandas as pd

df_corn = pd.read_csv('bitcoin_resource_data/corn_usd_daily_data.csv')
df_corn
```

```python
df_corn.isnull().sum()
```

```python
df_corn['거래량'] = df_corn['거래량'].apply(lambda x: float(x[:-1])*1000 if pd.notnull(x) else 0)
df_corn['거래량'] = df_corn['거래량'].apply(lambda x: round(x, 2))
df_corn['변동 %'] = df_corn['변동 %'].apply(lambda x: float(x[:-1]))
df_corn['날짜'] = pd.to_datetime(df_corn['날짜'])
```

```python
df_corn.dtypes
```

```python
import pandas as pd

df_gold = pd.read_csv('bitcoin_resource_data/gold_usd_daily_data.csv')
df_gold
```

```python
df_gold.isnull().sum()
```

```python
df_gold['거래량'].unique()
```

```python
df_gold['거래량'] = df_gold['거래량'].apply(lambda x: 0)
df_gold['거래량'] = df_gold['거래량'].apply(lambda x: round(x, 2))
df_gold['변동 %'] = df_gold['변동 %'].apply(lambda x: float(x[:-1]))
df_gold['날짜'] = pd.to_datetime(df_gold['날짜'])
```

```python
df_gold.astype({'종가':'float64', '시가':'float64', '고가':'float64', '저가':'float64'})
```

```python
for name in ['종가', '시가','고가','저가']:
    df_gold[name] = df_gold[name].str.replace(',','').astype('float64')
```

```python
df_gold
```

```python
df_gold.dtypes
```
