```python
from glob import glob

##      +   /* 
file_paths = glob('C:/Users/bluecom014/Downloads/299.,        /01-1./Training/01./TS_1.motion_m-_/*')

data = []
for i in file_paths:
    data.append(i.split('\\')[-1])


s = '_'.join(data[0].split('_')[:-1])
data2 = []
tmp = []
for idx, i in enumerate(data):
    a = i.split('_')
    b= a[-1].split('.')
    if '_'.join(a[:-1]) != s:
        data2.append(tmp)
        tmp = []
        tmp.append(i)
        s = '_'.join(a[:-1])
    else: tmp.append(i)

picture_data = []
cnt =1
for x in data2:
    num = 0
    for idx, y in enumerate(x):
        if num < int(y.split('.')[1][y.split('.')[1].find('e')+1:]):
            num = int(y.split('.')[1][y.split('.')[1].find('e')+1:])
            max_idx = idx
    picture_data.append(x[max_idx])
    cnt +=1

picture_data
```
