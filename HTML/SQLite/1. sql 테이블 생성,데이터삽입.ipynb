```python
#   (SQLite)
#    

import datetime
import sqlite3

#   
now = datetime.datetime.now()
print('now', now)

nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
print('nowDatetime', nowDatetime)

# sqlite3 
print('sqlite3.version : ', sqlite3.version)
print('sqlite3.sqlite_version', sqlite3.sqlite_version)

# DB & Autocommit
#  DB  

base_dir='C:/Users/bluecom014/Desktop/sesac_openai/HTML/sql/'
                                               #isolation_level=None    (Auto Commit)
conn = sqlite3.connect(base_dir+'eCommerce.db', isolation_level=None)  # :    
# DB()
# conn = sqlite3.connect(":memory:")

# Cursor
c = conn.cursor()
print('Cursor Type : ', type(c))

#  (Datatype : TEXT NUMERIC INTEGER REAL BLOB)
c.execute('CREATE TABLE IF NOT EXISTS  \
            (id INTEGER PRIMARY KEY,  text, email text, phone text, website text)')  # AUTOINCREMENT
```

```python
#  
c.execute("INSERT INTO  VALUES (1 ,'','@naver.com', '010-0000-0000', 'Kim.com')")
c.execute("INSERT INTO (id, , email, phone, website) VALUES (?, ?, ?, ?, ?)",
          (2, 'Park', 'Park@naver.com', '010-1111-1111', 'Park.com'))

# Many (, )
userList = (
    (3, 'Lee', 'Lee@naver.com', '010-2222-2222', 'Lee.com'),
    (4, 'Cho', 'Cho@naver.com', '010-3333-3333', 'Cho.com'),
    (5, 'Yoo', 'Yoo@naver.com', '010-4444-4444', 'Yoo.com')
)
c.executemany("INSERT INTO (id, , email, phone, website) VALUES (?, ?, ?, ?, ?)", userList)


#   
# print("users db deleted : ", conn.execute("delete from ").rowcount, "rows")

#  : isolation_level=None    (Auto Commit)
conn.commit()

# 
# conn.rollback()

#  
conn.close()
```
