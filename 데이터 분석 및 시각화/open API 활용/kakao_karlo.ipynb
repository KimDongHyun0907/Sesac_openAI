```python
from PyKakao import Karlo

key = 'Rest API key'
k = Karlo(service_key = key)

text = 'hyper realistic photo of cyberpunk sports car driving away'

imgDic = k.text_to_image(text, 1)
```

```python
img_path = imgDic['images'][0]['image']
img_path
```

```python
img_path = imgDic['images'][0]['image']

import requests
from PIL import Image
imgData = requests.get(img_path, stream=True)

with open('img_download.png', 'wb') as file:
    file.write(requests.get(img_path, stream=True).content)

img = Image.open(imgData.raw)
img
```
