네이버 쇼핑 api에서 '노트북', 제조사는 '삼성전자', 가격은 백만원 이하인 제품리스트 뽑기

```python
import os
import sys
import urllib.request
import pandas as pd
import json

client_id = ""
client_secret = ""
encText = urllib.parse.quote("노트북")
cnt = 0
dataList = []
for start in range(1, 31):
  url = "https://openapi.naver.com/v1/search/shop?query=" + encText+ "&start="+str(start) # json 결과
  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()
  if(rescode==200):
    response_body = response.read()
    json_data = json.loads(response_body.decode('utf-8'))
    for i in range(1, 10):
        dataList.append((json_data['items'][i]['title'], json_data['items'][i]['lprice'], json_data['items'][i]['maker']))
  else:
      print("Error Code:" + rescode)
```

```python
dataList = list(set(dataList))
```

```python
df = pd.DataFrame(dataList, columns=['제품명', '가격', '제조사'])
df
```

```python
df['가격'] = df['가격'].astype('int64')
```

```python
df.dtypes
```

```python
df['제품명'] = df['제품명'].apply(lambda x: x.replace('<b>', '').replace('</b>', ''))
```

네이버 쇼핑 api에서 'LG스마트 TV' 모든 자료 출력해서, 같은 제품별로 가장 최저가 검색해서 판매하는 쇼핑몰 및 링크 찾기

```python
import os
import sys
import urllib.request
import pandas as pd
import json

client_id = ""
client_secret = ""
encText = urllib.parse.quote("LG 스마트 TV")
cnt = 0
dataList = []
for start in range(1, 31):
  url = "https://openapi.naver.com/v1/search/shop?query=" + encText+ "&start="+str(start) # json 결과
  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()
  if(rescode==200):
    response_body = response.read()
    json_data = json.loads(response_body.decode('utf-8'))
    for i in range(1, 10):
        dataList.append((json_data['items'][i]['title'], 
                         json_data['items'][i]['lprice'], 
                         json_data['items'][i]['mallName'],
                         json_data['items'][i]['link']))
  else:
      print("Error Code:" + rescode)
```

```python
dataList = list(set(dataList))
```

```python
df = pd.DataFrame(dataList, columns=['제품명', '가격', '쇼핑몰', '링크'])
df
```

```python
df['제품명'] = df['제품명'].apply(lambda x: x.replace('<b>', '').replace('</b>', ''))
df['가격'] = df['가격'].astype('int64')
```

```python
df
```

```python
df = df[df['제품명'].apply(lambda x: 'LG' in x)]
df
```

```python
def fun(s):
    if s[-1] == ')' and s.find('(') != -1:
        num = s.find('(')
        return s[:num]
    else:
        return s

name = [fun(i) for i in df['제품명']]
name
```

```python
df.loc[:,'제품명'] = name
```

```python
tmp_df = df[['제품명','가격']].groupby('제품명').min()
tmp_df
```

```python
pd.merge(df, tmp_df, on = ['제품명', '가격'])
```
