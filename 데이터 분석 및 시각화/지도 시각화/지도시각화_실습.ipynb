```python
import folium
```

```python
# 위도
latitude = 37.394946
# 경도
longitude = 127.111104
```

```python
m = folium.Map(location=[latitude, longitude],
               zoom_start=17, 
               width=750, 
               height=500
              )
m
```

```python
folium.Marker([latitude, longitude],
              popup="판교역",
              tooltip="판교역 입구").add_to(m)
m
```

```python
folium.Marker([latitude, longitude],
              popup="판교역",
              tooltip="판교역 입구", 
              icon=folium.Icon('red', icon='star'),
             ).add_to(m)
m
```

```python
folium.Marker([latitude, longitude],
              popup='<iframe width="560" height="315" src="https://www.youtube.com/embed/dpwTOQri42s" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>',
              tooltip="판교역 입구").add_to(m)
m
```

```python
folium.CircleMarker([latitude, longitude],
                    color='tomato',
                    radius = 50, 
                    tooltip='판교역 상권').add_to(m)
m
```

```python
korea = pd.read_csv('../../data/csv 파일/korea.csv')
korea = korea.iloc[:-2,:]
korea = korea.iloc[::5, :]
```

```python
korea = korea.astype({'latitude':'float64'})
```

```python
map_korea = folium.Map(location=[36.5, 127.5], zoom_start=7)
map_korea
```

```python
for index, row in korea.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']], popup=row['docity'], icon=folium.Icon(color='blue', icon='info-sign')).add_to(map_korea)

map_korea
```

```python
map_korea = folium.Map(location=[36.5, 127.5], zoom_start=7)

for index, row in korea.iterrows():
    folium.CircleMarker(location=[row['latitude'], row['longitude']], popup=row['docity'], radius=5).add_to(map_korea)

map_korea
```

```python
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

m
```
