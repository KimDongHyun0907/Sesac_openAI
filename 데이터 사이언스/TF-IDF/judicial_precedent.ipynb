```python
def reason_content(x):
    if x.find('') != -1:
        return x.split('')[1]
    elif x.find(']') != -1:
        return x.split(']')[1]
    else:
        return x

df['_'] = df[''].apply(lambda x: reason_content(x))
```

```python
okt = Okt()

def make_token(x):
    words_noun = str()
    for word, tag in okt.pos(x):
        if tag in ['Noun','Adjective']:
            words_noun += word+' '
    return ' '.join(list(set(words_noun.split())))
```

```python
df_token[''] = df['_'].apply(lambda x: make_token(x))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math as math
import time 
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.cluster import MiniBatchKMeans
```

```python
print(plt.style.available)
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [14,14]
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
text_content = new_df['']
vector = TfidfVectorizer(max_df=0.4,  #  40%   
                             min_df=1,  #  1    
                             lowercase=True,
                             use_idf=True, 
                             norm=u'l2',  #  norm (L2 norm)
                             smooth_idf=True
                            )
tfidf = vector.fit_transform(text_content)
```

```python
k = 200  #   
kmeans = MiniBatchKMeans(n_clusters = k, n_init=3)  # MiniBatchKMeans  
kmeans.fit(tfidf)  # TF-IDF   ,  
centers = kmeans.cluster_centers_.argsort()[:,::-1]  #   TF-IDF   index  
terms = vector.get_feature_names_out()  #    
request_transform = vector.transform(new_df[''])  #     TF-IDF  

new_df['cluster'] = kmeans.predict(request_transform)  #    . cluster  
new_df['cluster'].value_counts().head()  #     
```

```python
def find_similar(tfidf_matrix, index, top_n = 5):
    cosine_similarities = linear_kernel(tfidf_matrix[index:index+1], tfidf_matrix).flatten()  #   
    related_docs_indices = [i for i in cosine_similarities.argsort()[::-1] if i != index]  #   index 
    return [index for index in related_docs_indices][0:top_n]  #    index 
```

```python
G = nx.Graph(label="judical")
start_time = time.time()
for i, rowi in new_df.iterrows():
    G.add_node(rowi[''], key=rowi[''], label='judical', jtype=rowi[''], court = rowi[''])

    indices = find_similar(tfidf, i, top_n = 5)

    snode = "SIM("+rowi[''][:15].strip()+")"
    G.add_node(snode, label="SIMILAR")
    G.add_edge(rowi[''], snode, label="SIMILARITY")

    for element in indices:
        G.add_edge(snode, new_df[''].loc[element], label='SIMILARITY')

print(" finish -- {} seconds --".format(time.time() - start_time))
```

```python
def get_all_adj_nodes(list_in):
    sub_graph=set()  # sub_graph 
    
    #   
    for m in list_in:
        sub_graph.add(m)
        for e in G.neighbors(m):        
                sub_graph.add(e)
    return list(sub_graph)
```

```python
def draw_sub_graph(sub_graph):
    subgraph = G.subgraph(sub_graph)  # sub_graph 
    colors=[]

    #      
    for e in subgraph.nodes():
        if G.nodes[e]['label']=="judical":
            colors.append('red') 
        elif G.nodes[e]['label']=="SIMILAR":
            colors.append('orange')    

    #  
    nx.draw(subgraph, with_labels=True, font_weight='bold',node_color=colors, font_family='Malgun Gothic')
    
    #  
    plt.show()
```

```python
list_in=["",""]
sub_graph = get_all_adj_nodes(list_in)
draw_sub_graph(sub_graph)
```

```python
def get_recommendation(root):  # root  
    commons_dict = {}
    for e in G.neighbors(root):  #    
        for e2 in G.neighbors(e):  #      
            if e2==root:  #    
                continue
            if G.nodes[e2]['label']=="judical":  #      
                commons = commons_dict.get(e2)  # commons :     
                if commons==None:  #    
                    #   
                    # e2 :      
                    # [e] :      
                    commons_dict.update({e2 : [e]}) 
                else:  #    
                    commons.append(e)  #     
                    # e2 :   
                    # commons :        
                    commons_dict.update({e2 : commons})
    
    movies=[]  #  
    weight=[]  #  
    for key, values in commons_dict.items():
        w=0.0  # 
        
        #          
        for e in values:
            w=w+1/math.log(G.degree(e))  #   (   )   .
        movies.append(key)
        weight.append(w)
    
    #      Series 
    #   index,  value
    result = pd.Series(data=np.array(weight),index=movies)

    #    
    result.sort_values(inplace=True,ascending=False)        
    return result
```

```python
result = get_recommendation("")
print(result.head())
print('-'*50)
result = get_recommendation("")
print(result.head())
```
