```python
def reason_content(x):
    if x.find('유】') != -1:
        return x.split('유】')[1]
    elif x.find('유]') != -1:
        return x.split('유]')[1]
    else:
        return x

df['판례내용_이유'] = df['판례내용'].apply(lambda x: reason_content(x))
```

```python
okt = Okt()

def make_token(x):
    words_noun = str()
    for word, tag in okt.pos(x):
        if tag in ['Noun','Adjective']:
            words_noun += word+' '
    return ' '.join(list(set(words_noun.split())))
```

```python
df_token['토큰'] = df['판례내용_이유'].apply(lambda x: make_token(x))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math as math
import time 
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.cluster import MiniBatchKMeans
```

```python
print(plt.style.available)
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [14,14]
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
text_content = new_df['토큰']
vector = TfidfVectorizer(max_df=0.4,  # 문서의 40%를 넘는 단어 제외
                             min_df=1,  # 최소 1번 이상 나오는 단어 사용
                             lowercase=True,
                             use_idf=True, 
                             norm=u'l2',  # 유클리드 norm (L2 norm)
                             smooth_idf=True
                            )
tfidf = vector.fit_transform(text_content)
```

```python
k = 200  # 클러스터의 개수 설정
kmeans = MiniBatchKMeans(n_clusters = k, n_init=3)  # MiniBatchKMeans 모델 초기화
kmeans.fit(tfidf)  # TF-IDF 벡터로 모델 학습, 클러스터링 수행
centers = kmeans.cluster_centers_.argsort()[:,::-1]  # 클러스터 중심점을 TF-IDF가 큰 순서대로 index를 내림차순으로 반환
terms = vector.get_feature_names_out()  # 클러스터링된 단어 목록 추출
request_transform = vector.transform(new_df['토큰'])  # 데이터의 토큰의 각 텍스트를 TF-IDF 벡터로 변환

new_df['cluster'] = kmeans.predict(request_transform)  # 새로운 데이터 클러스터를 예측. cluster 열에 저장
new_df['cluster'].value_counts().head()  # 클러스터에 속한 데이터 개수 계산
```

```python
def find_similar(tfidf_matrix, index, top_n = 5):
    cosine_similarities = linear_kernel(tfidf_matrix[index:index+1], tfidf_matrix).flatten()  # 코사인 유사도 계산
    related_docs_indices = [i for i in cosine_similarities.argsort()[::-1] if i != index]  # 유사한 문서 index 찾기
    return [index for index in related_docs_indices][0:top_n]  # 가장 유사한 문서 index 반환
```

```python
G = nx.Graph(label="judical")
start_time = time.time()
for i, rowi in new_df.iterrows():
    G.add_node(rowi['사건명'], key=rowi['판례일련번호'], label='judical', jtype=rowi['사건종류명'], court = rowi['법원명'])

    indices = find_similar(tfidf, i, top_n = 5)

    snode = "SIM("+rowi['사건명'][:15].strip()+")"
    G.add_node(snode, label="SIMILAR")
    G.add_edge(rowi['사건명'], snode, label="SIMILARITY")

    for element in indices:
        G.add_edge(snode, new_df['사건명'].loc[element], label='SIMILARITY')

print(" finish -- {} seconds --".format(time.time() - start_time))
```

```python
def get_all_adj_nodes(list_in):
    sub_graph=set()  # sub_graph 초기화
    
    # 인접한 노드 추가
    for m in list_in:
        sub_graph.add(m)
        for e in G.neighbors(m):        
                sub_graph.add(e)
    return list(sub_graph)
```

```python
def draw_sub_graph(sub_graph):
    subgraph = G.subgraph(sub_graph)  # sub_graph 생성
    colors=[]

    # 각 노드에 라벨에 따라 색상 설정
    for e in subgraph.nodes():
        if G.nodes[e]['label']=="judical":
            colors.append('red') 
        elif G.nodes[e]['label']=="SIMILAR":
            colors.append('orange')    

    # 그래프 그리기
    nx.draw(subgraph, with_labels=True, font_weight='bold',node_color=colors, font_family='Malgun Gothic')
    
    # 그래프 출력
    plt.show()
```

```python
list_in=["손해배상","임대차보증금"]
sub_graph = get_all_adj_nodes(list_in)
draw_sub_graph(sub_graph)
```

```python
def get_recommendation(root):  # root는 영화 제목
    commons_dict = {}
    for e in G.neighbors(root):  # 영화와 인접한 노드 확인
        for e2 in G.neighbors(e):  # 각 인접 노드에 인접한 노드 확인
            if e2==root:  # 입력 받은 노드는 제외
                continue
            if G.nodes[e2]['label']=="judical":  # 인접한 노드가 영화인 경우에만 공통 이웃
                commons = commons_dict.get(e2)  # commons : 공통 이웃으로 연결된 영화 리스트
                if commons==None:  # 공통 이웃이 없는 경우
                    # 새로운 항목 추가
                    # e2 : 현재 공통 이웃이 없는 영화 제목
                    # [e] : 현재 공통 이웃으로 연결된 영화 리스트
                    commons_dict.update({e2 : [e]}) 
                else:  # 공통 이웃이 있는 경우
                    commons.append(e)  # 새로운 공통 이웃을 리스트에 추가
                    # e2 : 해당 영화 제목
                    # commons : 해당 영화와 공통 이웃으로 연결된 다른 영화들 리스트
                    commons_dict.update({e2 : commons})
    
    movies=[]  # 영화 제목
    weight=[]  # 영화 가중치
    for key, values in commons_dict.items():
        w=0.0  # 가중치
        
        # 해당 영화와 공통 이웃으로 연결된 각 영화에 대한 가중치 계산
        for e in values:
            w=w+1/math.log(G.degree(e))  # 공통 이웃의 차수(노드와 연결된 간선의 수)의 역수를 가중치에 더함.
        movies.append(key)
        weight.append(w)
    
    # 영화 제목과 해당하는 영화 가중치의 Series 객체
    # 영화 제목을 index, 가중치를 value
    result = pd.Series(data=np.array(weight),index=movies)

    # 가중치가 큰 순서대로 정렬
    result.sort_values(inplace=True,ascending=False)        
    return result
```

```python
result = get_recommendation("손해배상")
print(result.head())
print('-'*50)
result = get_recommendation("임대차보증금")
print(result.head())
```
