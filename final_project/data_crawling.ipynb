```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time, random
from bs4 import BeautifulSoup

url='https://www.dcinside.com/'
comments=[]

options = Options()
options.add_experimental_option('detach', True)
options.add_argument('--disable-blink-features-AutomationControlled')

driver=webdriver.Chrome(options=options)

driver.get(url)
time.sleep(random.uniform(3, 6))
driver.get('https://gall.dcinside.com/')
time.sleep(random.uniform(3, 6))
driver.get('https://gall.dcinside.com/board/lists/?id=stock_new2')
time.sleep(random.uniform(3, 6))
for page_num in range(5, 101):
    driver.get('https://gall.dcinside.com/board/lists/?id=stock_new2&page='+str(page_num)+'&exception_mode=recommend')
    time.sleep(random.uniform(3, 6))
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    for num in soup.select('.gall_num'):
        if num.text.isdigit():
            driver.get('https://gall.dcinside.com/board/view/?id=stock_new2&no='+num.text+'&exception_mode=recommend&page='+str(page_num))
            time.sleep(random.uniform(3, 6))
            html = driver.page_source
            soup2 = BeautifulSoup(html, 'html.parser')
            if len(soup2.select('.comment_box > .cmt_list > .ub-content > div > div > p')) == 0:
                pass
            else:
                for comment in soup2.select('.comment_box > .cmt_list > .ub-content > div > div > p'):
                    comments.append(comment.text)
                    print(page_num, num.text, comment.text)
        driver.get('https://gall.dcinside.com/board/lists/?id=stock_new2&page='+str(page_num)+'&exception_mode=recommend')
        time.sleep(random.uniform(3, 6))
```

```python
import pandas as pd

df=pd.DataFrame(columns=['문장'])
df['문장'] = comments
# df.to_csv('comments_data.csv', encoding='cp949')
```
