```python
import pandas as pd
file_path = 'korean-hate-speech/labeled/train.tsv'
df_khs = pd.read_csv(file_path, sep='\t')
df_khs
```

```python
df_khs.groupby('hate').count()
```

```python
import numpy as np

df_khs['clean'] = np.where(df_khs['hate'] == 'none', 1, 0)
df_khs.drop(['contain_gender_bias', 'bias', 'hate'], axis=1, inplace=True)

df_khs
```

```python
df_khs.groupby('clean').count()
```

```python
from hangeul_onehot_vector import make_input_vector

pickle_data = pd.DataFrame(columns=['sentence', 'onehot_vector', 'clean'])
for comment in df_khs['comments']:
    data_vector = np.array(make_input_vector(comment))
    if data_vector.shape == (200, 127):
        target = df_khs.loc[df_khs['comments']==comment].values[0][1]
        pickle_data = pd.concat([pickle_data, pd.DataFrame({'sentence':[comment], 'onehot_vector':[data_vector], 'clean':[target]})], ignore_index=True)
```

```python
pickle_data
```

```python
# pickle íŒŒì¼ë¡œ ì €ì¥ (ìš©ëŸ‰ì´ ì»¤ì„œ gzip ì‚¬ìš©)
import pickle, gzip

with gzip.open('khs.pickle', 'wb') as f:
    pickle.dump(pickle_data, f)
```

```python
import json

with open('KOLD/data/kold_v1.json') as f:
    data_kold = json.load(f)

data_kold
```

```python
from hangeul_onehot_vector import make_input_vector

pickle_data = pd.DataFrame(columns=['sentence', 'onehot_vector', 'clean'])
for num in range(len(data_kold)):
    data_vector = np.array(make_input_vector(data_kold[num]['comment']))
    if data_vector.shape == (200, 127):
        target = 0 if data_kold[num]['OFF'] else 1
        pickle_data = pd.concat([pickle_data, pd.DataFrame({'sentence':[data_kold[num]['comment']], 'onehot_vector':[data_vector], 'clean':[target]})], ignore_index=True)
```
